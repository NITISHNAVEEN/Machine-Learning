```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("iris.csv")
df.head()
```

```python
df.Species.value_counts()
```

```python
df.isna().sum()
```

```python
df.drop(["Id"],axis =1,inplace = True)
```

```python
df["Species"] = df.Species.str.replace("Iris-","")
```

```python
df.head()
```

```python
df.info()
```

```python
sns.pairplot(data = df,hue = "Species",height = 2.75)
```

```python
df.head()
```

```python
df["Species_Num"] = df["Species"]
```

```python
df.head()
```

```python
df["Species_Num"] = df["Species_Num"].str.replace("setosa","1")
df["Species_Num"] = df["Species_Num"].str.replace("versicolor","2")
df["Species_Num"] = df["Species_Num"].str.replace("virginica","3")
```

```python
df.head()
```

```python
df.info()
```

```python
df["Species_Num"] = pd.to_numeric(df["Species_Num"],downcast = 'integer')
```

```python
df.info()
```

```python
plt.figure(figsize=(15,9))
df1 = df.drop(["Species"],axis = 1)
sns.heatmap(df1.corr(),cmap = "Spectral",annot = True)
```

```python
df.drop(['PetalLengthCm',"Species_Num"],axis = 1,inplace = True)
```

```python
df.head()
```

```python
x = df.drop(["Species"],axis = 1)
y = df.Species
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =  0.25,random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier()
```

```python
dt_model.fit(x_train,y_train)
```

```python
y_pred = dt_model.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print("Accuracy Score:",accuracy_score(y_test,y_pred))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_scores = cross_val_score(estimator = dt_model,X=x_train,y = y_train,cv = 10)
cv_scores.mean()
```

```python
cv_scores
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
cnf_matrix = confusion_matrix(y_test,y_pred)
sns.heatmap(cnf_matrix,annot = True,cmap = "YlGnBu")
plt.ylabel("Actual Label")
plt.xlabel("Predicted Label")
```

```python
print(classification_report(y_test,y_pred))
```

**OPTIMIZATION**

```python
from sklearn.model_selection import GridSearchCV
```

```python
dt_model2 = DecisionTreeClassifier()
```

```python
dt_params = {"criterion":["gini","entropy"],"max_depth":np.arange(2,22,2),"max_features":["auto","sqrt","log2"]}
dt_params
```

```python
dt_cv_model = GridSearchCV(estimator = dt_model2,param_grid= dt_params,cv = 10)
```

```python
dt_cv_model.fit(x_train,y_train)
```

```python
dt_cv_model.best_params_
```

```python
dt_tuned = DecisionTreeClassifier(criterion = "entropy",max_depth = 16,max_features = "log2")
```

```python
dt_tuned.fit(x_train,y_train)
```

```python
y_pred = dt_tuned.predict(x_test)
```

```python
print("Accuracy Score:",accuracy_score(y_test,y_pred))
```

```python
cv_scores = cross_val_score(estimator = dt_tuned,X=x_train,y = y_train,cv = 10)
cv_scores.mean()
```

**Hypermetric optimisation**

```python
rf_model2 = RandomForestClassifier()
```

```python
rf_params = {"n_estimators":[50,100,300],"max_depth":[3,5,7],"max_features":[2,4,6,8],"min_samples_split":[2,4,6]}
```

```python
rf_cv_model = GridSearchCV(rf_model2,rf_params,cv = 5,n_jobs = -1)
```

```python
rf_cv_model.fit(x_train,y_train)
```

```python
rf_cv_model.best_params_
```

```python
rf_tuned = RandomForestClassifier(max_depth = 3,max_features = 2,min_samples_split=2,n_estimators = 50)
```

```python
rf_tuned.fit(x_train,y_train)
```

```python
y_pred = rf_tuned.predict(x_test)
```
