**HOPKINS TEST**

```python
from pyclustertend import hopkins
from sklearn.preprocessing import scale
hopkins(scale(df),df.shape[0])
```

```python
from sklearn.cluster import KMeans
```

```python
k_means = KMeans(n_clusters = 2,random_state =42)
```

```python
k_means.fit(df)
```

```python
clusters = k_means.labels_
```

```python
clusters
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1],c = clusters,s = 75,cmap = "viridis")
```

```python
centroids = k_means.cluster_centers_
```

```python
centroids
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1],c = clusters,s = 75,cmap = "viridis")
plt.scatter(centroids[:,0],centroids[:,1],c = "red",s = 250,alpha = 0.50)
```

```python
rate = []
k = range(2,10)
for i in k:
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(df)
    rate.append(kmeans.inertia_)
```

```python
rate
```

```python
plt.plot(k,rate,"red")
plt.xlabel("k")
plt.title("Elbow")
```

```python
from yellowbrick.cluster import KElbowVisualizer
```

```python
kmeans_yellowbrick = KMeans()
graph = KElbowVisualizer(kmeans_yellowbrick,k=(2,10))
graph.fit(df)
graph.poof()
```

```python
new_kmeans = KMeans(n_clusters=4,random_state=42)
```

```python
new_kmeans.fit(df)
```

```python
labels = new_kmeans.labels_
```

```python
labels
```

```python
df["cluster_id"] = labels
```

```python
df.head()
```

**Silhoutte Score**

```python
from sklearn.metrics import silhouette_score
```

```python
print(f"Silhoutte Score(n=4):{silhouette_score(df,labels)}")
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1],c = labels,s = 75,cmap = "viridis")
plt.scatter(new_kmeans.cluster_centers_[:,0],new_kmeans.cluster_centers_[:,1],c = "red",s = 250,alpha = 0.50)
```
